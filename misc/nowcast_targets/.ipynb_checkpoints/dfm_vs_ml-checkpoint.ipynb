```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [14, 7]
```

```python
catalog = pd.read_csv("../helper/catalog.csv")
data = pd.read_csv("../output/2020-06-01_database_tf.csv", parse_dates=["date"])
ys = ["x_world.sa", "x_vol_world2.sa", "x_servs_world.sa"]
def gen_y(name):
    y = data.loc[:, ["date", name]]
    for i in range(1,len(y)-1):
        if y.date.dt.month[i] in [1, 4, 7, 10]:
            y.loc[i, name] = y.loc[i-1, name]
        elif y.date.dt.month[i] in [2, 5, 8, 11]:
            y.loc[i, name] = y.loc[i+1, name]
    return y
all_y = gen_y(ys[0])
```

```python
def gen_vintage(data, catalog, series):
    check_series = series
    if (catalog.code == series).sum() == 0:
        check_series = series[:-3]
    if (catalog.code == check_series).sum() > 0:
        lag = catalog.loc[catalog.code == check_series, "publication_lag"].values[0]
        return data[series].shift(lag)
vintage = data.drop(ys, axis=1)
for col in data.columns[1:]:
    vintage[col] = gen_vintage(data, catalog, col)
```
